```python
import crewai
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
```

```python
print(crewai.__version__)
```

Initializing the tool

```python
search_tool = SerperDevTool()
```

Defining the Agent

```python
researcher = Agent(
    role='Mangoes Research Analyst',
    goal='Analyze current trends related to Mango farming in Western Ghats of India.',
    backstory='You are an expert in identifying and summarizing the latest agricultural trends in India.',
    tools=[search_tool],
    verbose=True
)
```

Defining a task

```python
research_task = Task(
    description='Are Mangoes grown in Karnataka?',
    expected_output='Give at least 4 places where mangoes are grown in Karnataka',
    agent=researcher
)
```

Building the crew

```python
crew = Crew(
    agents=[researcher],
    tasks=[research_task],
    process=Process.sequential
)
```

Running the crew

```python
result = crew.kickoff()
print(result)
```
