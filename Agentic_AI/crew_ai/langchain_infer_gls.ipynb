```python
import pickle
import numpy as np 
import pandas as pd 
import os 
from pydantic import BaseModel
from langchain_groq import ChatGroq
from dotenv import load_dotenv
load_dotenv()
```

```python
model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.0,
    max_retries=2,
)

class GlassInfo(BaseModel):
    RI: float
    Na: float
    Mg: float
    Al: float
    Si: float
    K: float
    Ca: float
    Ba: float
    Fe: float

structured_llm = model.with_structured_output(GlassInfo)
recorded_conversation = """ I have a piece of glass with contents as follows: Aluminum-34, RI-78
                            sodium-40, silicon-30,calcium-3, barium-4, iron-3"""

result =  structured_llm.invoke(recorded_conversation)
print(result)
```
