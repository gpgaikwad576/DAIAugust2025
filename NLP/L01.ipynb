```python
sent = "They told that their ages are 26,31 and 29 respectively"
```

```python
#average of ages in this sentence
lst = list(map(int,re.findall(r"\d+",sent)))
sum(lst)/len(lst)
```

```python
words = sent.split()
```

```python
num = []
for word in words:
    if word.isdigit():
        num.append(int(word))
sum(num)/len(num)
```

```python
num = [int(word) for word in sent.split() if word.isdigit()]
sum(num)/len(num)
```

MUltiword expression(mwe) Tokenizer

```python
nsent = 'Chhatrapati Shivaji Maharaj was founder of Maratha Empire.'
word_tokenize(nsent)
```

```python
from nltk.tokenize import MWETokenizer
```

```python
tk = MWETokenizer(separator = ' ')
```

```python
tk.add_mwe(('Chhatrapati','Shivaji'),('Maratha','Empire'))
tk.tokenize(word_tokenize(nsent))
```
