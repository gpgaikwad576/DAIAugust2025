```python
cap = cv2.VideoCapture(r'datasets\image_dataset\images\robots.mp4')
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()


output_dir = 'datasets/image_dataset/images/output_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print('Created directory Successfully: ',output_dir)

cnt=1

while cap.isOpened():
    _,frame = cap.read()

    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    

    if _ :
        cv2.imshow('Video Frame',frame)
        cv2.imshow('gray video',gray)
        
        cv2.imwrite(f'datasets/image_dataset/images/output_images/frame{cnt}.png',frame)
        cnt+=1

        if cv2.waitKey(10)==27:
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
translated_img = cv2.warpAffine(img,T,(img.shape[1],img.shape[0]))
```

```python
translated_img = cv2.warpAffine(img,T,(600,600))
cv2.imshow('Translated Image',translated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
T = np.float32([[1,0,150],[0,1,75]])
translated_img = cv2.warpAffine(img,T,(img.shape[1],img.shape[0]))
cv2.imshow('Translated_image',translated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Image flip

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_hor = cv2.flip(img_rgb,0)
img_ver = cv2.flip(img_rgb,1)
img_dia = cv2.flip(img_rgb,-1)
```

```python
plt.figure(figsize = (16,9))
plt.subplot(141)
plt.imshow(img_rgb); plt.title('Original')
plt.subplot(142)
plt.imshow(img_hor); plt.title('Horizontal')
plt.subplot(143)
plt.imshow(img_ver); plt.title('Vertical')
plt.subplot(144)
plt.imshow(img_dia); plt.title('Diagonalal')
```
