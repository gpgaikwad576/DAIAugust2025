```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python

lines = [
    "It was a nice rainy day.",
    "The things were so beautiful in his point",
    "When your focus is clear, you won.",
    "Many many returns of the day"
]
```

```python
tokenizer =Tokenizer()
tokenizer.fit_on_texts(lines)
tokenizer.word_docs
```

```python
tokenizer.index_docs
```

```python
tokenizer.word_index
```

```python
mat =tokenizer.texts_to_matrix(lines)
```

```python
mat
```

```python
seq = tokenizer.texts_to_sequences(lines)
```

```python
seq
```

```python
padded = pad_sequences(seq,maxlen=10,padding="post")
padded
```

```python
padded = pad_sequences(seq,maxlen=10,padding="pre")
padded
```

```python
maxi=0
for i in seq:
    maxi=max(maxi,len(i))
maxi
```

```python
padded = pad_sequences(seq,maxlen=maxi,padding="post")
padded
```
