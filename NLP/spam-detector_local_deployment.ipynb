```python
from tkinter import *
import joblib
from nltk import word_tokenize
```

```python
top = Tk()
top.title("SPAM Message Detector")
top.geometry("800x500")


from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
ps = PorterStemmer()
swords = stopwords.words('english')

def clean_text2(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4

classifier = joblib.load('classifier.model')
preprocessor = joblib.load('preprocessor.model')

def check():
    message = t.get("1.0",END)
    mes_new = preprocessor.transform([message])
    pred = classifier.predict(mes_new)
    if pred[0] == "ham":
        result.config(text = "Message is not SPAM!")
    else:
        result.config(text = "Message is SPAM!")

Label(top,text = "Enter your message here.").place(x = 100,y = 10)
t = Text(top)
t.place(x = 20,y =40)
b = Button(top,text = "Check message",command = check)
b.place(x = 200, y = 450)
result = Label(text = "Message is")
result.place(x = 400, y = 450)
top.mainloop()
```
