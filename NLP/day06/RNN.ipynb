```python
import numpy as np
import tensorflow
import keras
from keras.utils import to_categorical
```

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict(list(enumerate(alphabets)))
```

```python
int_to_char[6]
```

```python
dict((v,i) for i,v in enumerate(alphabets))
```

```python
char_to_int = dict((v,i) for i,v in enumerate(alphabets))
```

```python
char_to_int['X']
```

Prepare the dataset

```python
x = []
y = []
seq_length = 1

for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    print(seq_in, '-->',seq_out)
    x+=[[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
```

```python
len(x)
```

```python
x = np.reshape(x,(25,1,1))
x
```

```python
x_norm = x/25
```

```python
x_norm
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

```python
x_norm.shape
```

Build the model

```python
from keras.layers import SimpleRNN,Dense,Input
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Input((1,1)))
model.add(Dense(25,activation = 'relu'))
model.add(SimpleRNN(32))
model.add(Dense(26,activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

Train the model

```python
model.fit(x_norm,y_new,epochs = 500,batch_size = 1)
```

```python
model.evaluate(x_norm,y_new,batch_size = 1)
```

```python
new = 'M'
new = char_to_int[new]
new = np.reshape(new,(1,1,1,1))
new = new/25
pred = model.predict(new,verbose = False)
pred = np.argmax(pred)
int_to_char[pred]
```

`For seq_length = 3`

```python
x = []
y = []
seq_length = 3

for i in range(len(alphabets)-seq_length):
    seq_in = alphabets[i:i + seq_length]
    seq_out = alphabets[i+seq_length]
    print(seq_in, '-->',seq_out)
    x.append([char_to_int[char] for char in seq_in])
    y += [char_to_int[seq_out]]
```

```python
len(x)
```

```python
x = np.reshape(x,(23,seq_length,1))
x
```

```python
x_norm = x/25
```

```python
x_norm
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

```python
x_norm.shape
```

```python
model = Sequential()
model.add(Input((seq_length,1)))
model.add(Dense(25,activation = 'relu'))
model.add(SimpleRNN(32))
model.add(Dense(26,activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
model.fit(x_norm,y_new,epochs = 1000,batch_size = 1)
```

```python
new = 'WXY'
val = np.array([[char_to_int[x]] for x in new])
val = val.reshape(1,3,1)
val = val/25
result = model.predict(val,verbose = False)
result = np.argmax(result)
print(new,'-->',int_to_char[result])
```
