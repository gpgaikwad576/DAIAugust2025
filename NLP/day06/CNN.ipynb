```python
import string, re, nltk, os
from nltk.corpus import stopwords
from tensorflow.keras.preprocessing.text import Tokenizer
from nltk.tokenize import word_tokenize
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
```

```python
#load the doc in memory
def load_doc(filename):
    f = open(filename)
    text = f.read()
    f.close()
    return text
```

```python
swords=stopwords.words("english")
```

```python
# Function to clean the doc
def clean_doc(doc):
    tokens = word_tokenize(doc)
    tokens = [token for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if token not in swords]
    tokens = [token for token in tokens if len(tokens) > 1]

    return tokens
```

```python
f=open(r"D:\NLPDS\Datasets\vocab.txt")
vocab=f.read()
vocab
```

```python
# load the file, Clean the data and return lines

def doc_to_line(filename):
    doc=load_doc(filename)
    tokens = clean_doc(doc)
    tokens = [token for token in tokens if token in vocab]
    return " ".join(tokens)
```

```python
def process_train(directory):
    documents = []
    for filename in os.listdir(directory):
        if not filename.startswith("cv9"):
            path = directory + "\\" + filename
            docs =load_doc(path)
            tokens = clean_doc(docs)
            documents.append(tokens)
        
    return documents
```

```python
def process_docs(directory,is_train):
    documents = []
    for filename in os.listdir(directory):
        if is_train and filename.startswith("cv9"):
            continue
        if not is_train and not filename.startswith("cv9"):
            continue
        path = directory + "\\" + filename
        docs =load_doc(path)
        tokens = clean_doc(docs)
        documents.append(tokens)
        
    return documents
```

```python
def load_data(is_train):
    neg=process_docs(r"D:\NLPDS\Datasets\review_polarity\txt_sentoken\neg",is_train)
    pos=process_docs(r"D:\NLPDS\Datasets\review_polarity\txt_sentoken\pos",is_train)
    docs=neg+pos
    labels=[0 for i in range((len(neg)))]+[1 for i in range((len(pos)))]
    return docs,labels
```

## **Data Preparation**

```python
def create_tokenizer(lines):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(lines)
    return tokenizer
```

```python
tokenizer =  create_tokenizer(train_data)
```

```python
# Encode the training data

X_train = tokenizer.texts_to_matrix(train_data)
```

```python
def encode_docs(tokenizera,max_lengtha,docs):
    encoded = tokenizera.texts_to_sequences(docs)
    padded = pad_sequences(encoded,maxlen = max_lengtha,padding = 'post')
    return padded
```

## **Build the model**

```python
from keras.models import Sequential
from keras.layers import Dense,Input,Conv1D,MaxPool1D,Flatten,Embedding
from keras.utils import plot_model
```

```python
vocab_size = len(vocab)
```

```python
from sklearn import metrics


def define_model(max_length):
    model=Sequential()
    model.add(Embedding(vocab_size,100,input_length = max_length))
    model.add(Conv1D(filters = 16,kernel_size = 5, activation = 'relu'))
    model.add(MaxPool1D(pool_size = 2))
    model.add(Flatten())
    model.add(Dense(10, activation="relu"))
    model.add(Dense(1, activation ="sigmoid"))
    model.compile(optimizer = "adam",loss="binary_crossentropy",metrics=["accuracy"])
    return model
```

```python
model =define_model(X_train.shape[1])
```

```python
plot_model(model,show_layer_activations=True,show_layer_names=True,show_shapes=True)
```

## **Train the model**

```python
history = model.fit(X_train,np.array(train_labels),epochs=10,batch_size=5)
```
