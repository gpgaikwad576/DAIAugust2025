Extract specific POS from the word

```python
verbs = set()
nouns = set()
adjts = set()
```

```python
#extract the synonyms sound
for word in wordnet.synsets('sound'):
    pos_code = word.pos()
    if(pos_code == wordnet.NOUN):
        nouns.add(word.name())
    if(pos_code == wordnet.VERB):
        verbs.add(word.name())
    if(pos_code == wordnet.ADJ):
        adjts.add(word.name())
```

```python

print(nouns)
print(verbs)
adjts
```

```python
for word in wordnet.synsets('sound', pos = 'a'):
    adjts.add(word.lemmas()[0].name())
```

```python
adjts
```

Retrive the antonyms

```python
antos = set()
for word in wordnet.synsets('possible'):
    for lemma in word.lemmas():
        if lemma.antonyms():
            antos.add(lemma.antonyms()[0].name())
```

```python
swords = stopwords.words('english')
ps = PorterStemmer()
wnl = WordNetLemmatizer()
```

```python
def clean_text(sent):
    tokens1 = word_tokenize(sent)
    tokens2 = [token for token in tokens1 if token.isalpha()]
    tokens3 = [ps.stem(token.lower()) for token in tokens2 if token.lower() not in swords]
    return tokens3
```
