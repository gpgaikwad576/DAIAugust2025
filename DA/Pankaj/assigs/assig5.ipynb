```python
from matplotlib import pyplot  as plt
import seaborn as sns
import pandas as pd
import numpy as np
from io import StringIO
```

```python
sns.set_style("white")
```

```python
# 1. Write a Python programming to display a bar chart of the popularity of
# programming Languages
data = '''
proglangs,popularity
Java,22.2
Python,17.6
PHP,8.8
JavaScript,8
C#,7.7
C++,6.7
'''
df = pd.read_csv(StringIO(data))
```

```python
sns.barplot(df,x = 'proglangs',y = 'popularity')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
```

```python
# 2. Write a Python programming to display a horizontal bar chart of the
# popularity of programming Languages.
lang_ordered = df.sort_values('popularity',ascending = True)['proglangs'].tolist()
sns.barplot(df,x = 'popularity',y = 'proglangs',order = lang_ordered)
plt.grid(which = 'major',ls = '-',lw = '1.5')
plt.grid(which = 'minor',ls = '-',lw = 0.5)
plt.minorticks_on()
```

```python
# 3. (inverted) Write a Python programming to display a bar chart of the popularity of
# programming Languages. Use uniform color.
sns.barplot(df,x = 'proglangs',y = 'popularity')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
plt.gca().invert_yaxis()
```

```python
# 4. Write a Python programming to display a bar chart of the popularity of
# programming Languages. Use different color for each bar.
sns.barplot(df,x = 'proglangs',y = 'popularity',hue = 'proglangs',palette = 'viridis')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
```

```python
# 5. Write a Python programming to display a bar chart of the popularity of
# programming Languages. Attach a text label above each bar displaying its
# popularity (float value).
sns.barplot(df,x = 'proglangs',y = 'popularity')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
for i,v in enumerate(df['popularity']):
    plt.text(i,v+0.5,f'{v}%', ha= 'center',va = 'bottom', fontweight = 'bold')
plt.title('Poularity of Programming languages')
```

```python
# 6. Write a Python programming to display a bar chart of the popularity of
# programming Languages. Make blue border to each bar.
sns.barplot(df,x = 'proglangs',y = 'popularity',edgecolor = 'blue', linewidth = 2)
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
plt.title('Poularity of Programming languages')
```
