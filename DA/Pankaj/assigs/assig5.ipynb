```python
from matplotlib import pyplot  as plt
import seaborn as sns
import pandas as pd
import numpy as np
from io import StringIO
```

```python
sns.set_style("white")
```

```python
# 1. Write a Python programming to display a bar chart of the popularity of
# programming Languages
data = '''
proglangs,popularity
Java,22.2
Python,17.6
PHP,8.8
JavaScript,8
C#,7.7
C++,6.7
'''
df = pd.read_csv(StringIO(data))
```

```python
sns.barplot(df,x = 'proglangs',y = 'popularity')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
```

```python
# 2. Write a Python programming to display a horizontal bar chart of the
# popularity of programming Languages.
lang_ordered = df.sort_values('popularity',ascending = True)['proglangs'].tolist()
sns.barplot(df,x = 'popularity',y = 'proglangs',order = lang_ordered)
plt.grid(which = 'major',ls = '-',lw = '1.5')
plt.grid(which = 'minor',ls = '-',lw = 0.5)
plt.minorticks_on()
```

```python
# 3. (inverted) Write a Python programming to display a bar chart of the popularity of
# programming Languages. Use uniform color.
sns.barplot(df,x = 'proglangs',y = 'popularity')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
plt.gca().invert_yaxis()
```

```python
# 4. Write a Python programming to display a bar chart of the popularity of
# programming Languages. Use different color for each bar.
sns.barplot(df,x = 'proglangs',y = 'popularity',hue = 'proglangs',palette = 'viridis')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
```

```python
# 5. Write a Python programming to display a bar chart of the popularity of
# programming Languages. Attach a text label above each bar displaying its
# popularity (float value).
sns.barplot(df,x = 'proglangs',y = 'popularity')
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
for i,v in enumerate(df['popularity']):
    plt.text(i,v+0.5,f'{v}%', ha= 'center',va = 'bottom', fontweight = 'bold')
plt.title('Poularity of Programming languages')
```

```python
# 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women
Means_men = [22, 30, 35, 35, 26]
Means_women = [25, 32, 30, 35, 29]
groups = ['G1','G2','G3','G4','G5']
df= pd.DataFrame({'groups':groups*2,'Gender':['Men']*len(Means_men)+['Women']*len(Means_women),'Scores':Means_men + Means_women})
sns.barplot(data = df,x='groups',y = 'Scores',hue='Gender',palette = ['g','r'])
plt.grid(which ='major',ls = '-',lw = '1.5')
plt.grid(which = 'minor' ,ls = '--',lw ='0.5')
plt.minorticks_on()
plt.title('Popularity of Programming languages')
```

```python
# 11. Write a Python program to create bar plot from a DataFrame.
data = '''
a,b,c,d,e
4,8,5,7,6
2,3,4,2,6
4,7,4,7,8
2,6,4,8,6
2,4,3,3,2
'''
df = pd.read_csv(StringIO(data))
groups = ['G1','G2','G3','G4','G5']
df2 = pd.DataFrame({'groups':groups*5,'vars':['a']*len(df.a)+['b']*len(df.b)+['c']*len(df.c)+['d']*len(df.d)+['e']*len(df.e),'vals':df.values.ravel()})
sns.barplot(data = df2,x='groups',y='vals',hue='vars',palette = ['r','g','b','y','black'])
plt.title('a,b,c,d,e values')
plt.plot()
```

```python
# 12. Write a Python program to create bar plots with error bars on the same
# figure.
partis = [1,2,3,4]
vel = [30,45,40,50]
std_vel = [5, 6, 7, 8]
sns.barplot(x= partis, y =vel)
plt.errorbar(x = np.arange(len(partis)), y = vel,yerr = std_vel,fmt= 'none',color = 'black', capsize = 6)
plt.title('bars with velocity')
```

```python
# 13. Write a Python program to create bar plots with errorbars on the same
# figure. Attach a text label above each bar displaying men means (integer
# value)
vel = [30,45,40,50]
std_vel = [5, 6, 7, 8]
sns.barplot(x=np.arange(4),y =vel)
plt.errorbar(x=np.arange(4),y = vel,yerr = std_vel,fmt = 'none',color = 'black', capsize =6)
for i,v in enumerate(vel):
    plt.text(i,v+std_vel[i]+0.5,f'{v}%', ha = 'center',va = 'bottom',fontweight = 'bold')
```

```python
# 14. Write a Python program to create a stacked bar plot with error bars.
# Note: Use bottom to stack the women?s bars on top of the men?s bars
# Means (men) = (22, 30, 35, 35, 26)
# Means (women) = (25, 32, 30, 35, 29)
# Men Standard deviation = (4, 3, 4, 1, 5)
# Women Standard deviation = (3, 5, 2, 3, 3)
categories = ['A', 'B', 'C', 'D', 'E']
categories2 = ['A', 'B', 'C', 'D', 'E']
men_means = [22, 30, 35, 35, 26]
women_means = [25, 32, 30, 35, 29]
men_std = [4, 3, 4, 1, 5]
women_std = [3, 5, 2, 3, 3]
# plt.figure(figsize =(10,6))
bars1 = plt.bar(categories,men_means,yerr = men_std, capsize = 5,label = 'Men',color = 'blue',alpha = 0.7)
bars2 = plt.bar(categories2,women_means,yerr = women_std,capsize = 5, label = 'Women',color = 'red',alpha = 0.7)
plt.title('Stacked bar plot with Error bar')
plt.legend()
plt.show()
```
