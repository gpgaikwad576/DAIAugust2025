```python
import matplotlib.pyplot as plt

languages = [['Language','Science','Math'],
             ['Science','Math','Language'],
             ['Math','Language','Science']]
numbers = [{'Language':75, 'Science':88, 'Math':96},
           {'Language':71, 'Science':95, 'Math':92},
           {'Language':75, 'Science':90, 'Math':89}]

colors = {'Language': 'skyblue', 'Science': 'lightcoral', 'Math': 'lightgreen'}
students = ['Student 1', 'Student 2', 'Student 3']

fig, ax = plt.subplots(figsize=(10, 6))

for i, (order, data) in enumerate(zip(languages, numbers)):
    left = 0
    for subject in order:
        ax.barh(students[i], data[subject], left=left,
                color=colors[subject], label=subject if i == 0 else "")
        left += data[subject]

plt.xlabel('Total Score')
plt.title('Student Scores - Stacked Horizontal Bars')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 16. Write a Python program to create stack bar plot and add label to each
# section
people = ['G1','G2','G3','G4','G5','G6','G7','G8']
data = [[ 3.40022085, 7.70632498, 6.4097905, 10.51648577, 7.5330039,
7.1123587, 12.77792868, 3.44773477],
[ 11.24811149, 5.03778215, 6.65808464, 12.32220677, 7.45964195,
6.79685302, 7.24578743, 3.69371847],
[ 3.94253354, 4.74763549, 11.73529246, 4.6465543, 12.9952182,
4.63832778, 11.16849999, 8.56883433],
[ 4.24409799, 12.71746612, 11.3772169, 9.00514257, 10.47084185,
10.97567589, 3.98287652, 8.80552122]]

df = pd.DataFrame(data).T
df.index =people
df.columns = ['cat1','cat2','cat3','cat4']

colors = {'cat1':'skyblue','cat2':'black','cat3':'lightgreen','cat4':'red'}
fig,ax = plt.subplots(figsize = (10,6))
for group in df.index:
    left =0
    for cat in df.columns:
        value = df.loc[group,cat]
        ax.barh(group,value,left = left, color =colors[cat],label = cat if group =='G1' else "")
        left+=value
plt.xlabel('Category Scores')
plt.ylabel('Groups')
plt.title('Groups and their category scores')
plt.legend()
plt.tight_layout()


for group_idx, group in enumerate(df.index):
    left = 0
    for cat in df.columns:
        value = df.loc[group, cat]
        # Position label in the middle of each segment
        ax.text(left + value/2, group_idx, f'{value:.1f}',
                ha='center', va='center', fontweight='bold', color='white')
        left += value
```

```python
# 17. Write a Python program to add textures (black and white) to bars and
# wedges
import matplotlib.pyplot as plt
import numpy as np

# Data for bar plot
categories = ['A', 'B', 'C', 'D']

# Textures/patterns
textures = ['/', '\\', '*', 'O']

fig, ax = plt.subplots(figsize=(10, 6))

# Create bars with textures
bars = ax.bar(categories, 3, color='white', edgecolor='black', linewidth=1.5)

# Apply textures
for bar, texture in zip(bars, textures):
    bar.set_hatch(texture * 3)  # Repeat pattern for density

plt.title('Bar Plot with Textures')
plt.ylabel('Values')
plt.show()
```
