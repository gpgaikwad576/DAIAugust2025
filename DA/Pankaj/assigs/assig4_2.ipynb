```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
#### 1. Write a Python program to draw a line with suitable label in the x axis, y
#### axis and a title
x= [0,5,10,15,20,25,30]
y = x
plt.plot(x,y)
plt.title('simple line')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True)
plt.show()
```

```python
# 3. Write a Python program to draw a line using given axis values taken from a
# text file, with suitable label in the x axis, y axis and a title.
# Test Data:
# test.txt
# 1 2
# 2 4
# 3 1
data = np.loadtxt('file.txt')
x = data[:,0]
y = data[:,1]
plt.plot(x,y,'g-',linewidth =2)
plt.title('Sample graph')
plt.xlabel('x-axis')
plt.ylabel('y')
plt.show()
```

```python
# 4. Write a Python program to draw line charts of the financial data of Alphabet
# Inc. between October 3, 2016 to October 7, 2016.
df = pd.read_csv('fdata.csv')
df['Date'] = pd.to_datetime(df['Date'],format = '%m-%d-%y')
plt.figure(figsize = (10,6))
plt.plot(df['Date'],df['Open'],label ='Open')
plt.plot(df['Date'],df['High'],label = 'High')
plt.plot(df['Date'],df['Low'],label = 'Low')
plt.plot(df['Date'],df['Close'],label = 'Close')

plt.title('Alphabet Inc. Stock Prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6. Write a Python program to plot two or more lines with legends, different
# widths and colors.
x = [10,15,20,25,30]
y1 = [20,30,40,26,10]
y2 = [40,30,20,26,30]
plt.plot(x,y1, label ='y1',linewidth = 4, color = 'Red')
plt.plot(x,y2,label = 'y2',linewidth = 2, color = 'Yellow')
plt.legend()
plt.title('Sample graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 8. Write a Python program to plot two or more lines and set the line markers.
x = [10,15,20,25,30]
y1 = [20,30,40,26,10]
y2 = [40,30,20,26,30]
plt.plot(x,y1, 'ro:',label ='y1',markersize=6, markerfacecolor = 'Blue')
plt.plot(x,y2, 'rs--',label = 'y2', markersize = 6, markerfacecolor = 'Blue')
plt.legend()
plt.title('Sample graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 14. Write a Python program to display the grid and draw line charts of the
# closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016.
# Customized the grid lines with rendering with a larger grid (major grid) and a
# smaller grid (minor grid).Turn on the grid but turn off ticks.
# Sample data
data ='''
Date,Close
03-10-16,772.559998
04-10-16,776.429993
05-10-16,776.469971
06-10-16,776.859985
07-10-16,775.080017
'''
df = pd.read_csv(StringIO(data))
df['Date'] = pd.to_datetime(df['Date'],format = '%d-%m-%y')
plt.plot(df['Date'],df['Close'])
plt.scatter(df['Date'],df['Close'],)
plt.title('sample graph')
plt.xlabel('x')
plt.xticks(rotation = 45)
plt.ylabel('y')
plt.grid(which = 'major', ls = '-',lw=1.5,c='b')
plt.grid(which = 'minor', ls = '--',lw=0.5,c='b')
plt.minorticks_on()
plt.tick_params(which = 'both',bottom = False,left = False, top = False, right = False)
plt.show()
```
