```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import os 
os.chdir("C:/Python/Datasets")
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv("milk.csv", index_col=0)
```

```python
clf = IsolationForest(contamination=0.05, random_state=25)
clf.fit(df)
predictions = clf.predict(df)
dict( zip(list(df.index), predictions) )
```

```python
nut = pd.read_csv("nutrient.csv", index_col=0)
clf = IsolationForest(random_state=25)
clf.fit(nut)
predictions = clf.predict(nut)
dict( zip(list(nut.index), predictions) )
```
