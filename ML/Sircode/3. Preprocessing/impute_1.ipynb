#### Chemical Process

```python
chem = pd.read_csv("C:/Python/Cases/Chemical_Process_Data/ChemicalProcess.csv")
y = chem['Yield']
X = chem.drop('Yield', axis=1)
```

```python
y.isnull().sum()
```

```python
X.isnull().sum().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=25)
```

```python
X_train.isnull().sum().sum()
```

```python
X_test.isnull().sum().sum()
```

```python
imputer = SimpleImputer(strategy='mean').set_output(transform='pandas')
imputer.fit(X_train)
X_trn_imp = imputer.transform(X_train)
X_trn_imp.isnull().sum().sum()
```

```python
X_tst_imp = imputer.transform(X_test)
X_tst_imp.isnull().sum().sum()
```

```python
lr = LinearRegression()
lr.fit(X_trn_imp, y_train)
y_pred = lr.predict(X_tst_imp)
mean_absolute_error(y_test, y_pred)
```

Median Imputation

```python
imputer = SimpleImputer(strategy='median').set_output(transform='pandas')
X_trn_imp = imputer.fit_transform(X_train)
X_tst_imp = imputer.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_imp, y_train)
y_pred = lr.predict(X_tst_imp)
mean_absolute_error(y_test, y_pred)
```

#### Iterative Imputer

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
imputer = IterativeImputer(initial_strategy='mean')
X_trn_imp = imputer.fit_transform(X_train)
X_tst_imp = imputer.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_imp, y_train)
y_pred = lr.predict(X_tst_imp)
mean_absolute_error(y_test, y_pred)
```

```python
imputer = IterativeImputer(initial_strategy='median')
X_trn_imp = imputer.fit_transform(X_train)
X_tst_imp = imputer.transform(X_test)
lr = LinearRegression()
lr.fit(X_trn_imp, y_train)
y_pred = lr.predict(X_tst_imp)
mean_absolute_error(y_test, y_pred)
```
