```python
import numpy as np
import pandas as pd
from sklearn.metrics import log_loss
```

```python
hr = pd.read_csv("C:/Python/Cases/human-resources-analytics/HR_comma_sep.csv")
x = hr['satisfaction_level'].values
y = hr['left'].values
```

```python
hr
```

```python
eta, w, b = 0.1, 0.1, 0.1
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
ycap = sigmoid(w*x + b)
L = log_loss( y , ycap)
L
```

```python
dw = -np.mean((y-ycap)*x)
db = -np.mean(y-ycap)
dw, db
```

```python
eta = 0.4
b, w = 0.1, 0.1
for i in range(10000):
    ycap = sigmoid(w*x + b)
    L = log_loss( y , ycap)
    if i%100 == 0:
        print(f"w={w}, b={b}, Loss={L}")
    dw = -np.mean((y-ycap)*x)
    db = -np.mean(y-ycap)
    new_w = w - eta*dw 
    new_b = b - eta*db 
    w,b = new_w, new_b

```

Taking two features

```python
x1 = hr['satisfaction_level'].values
x2 = hr['number_project'].values
y = hr['left'].values
```

```python
eta = 0.4
b, w1, w2 = 0.1, 0.1, 0.1
for i in range(10000):
    ycap = sigmoid(w1*x1 + w2*x2 +b)
    L = log_loss( y , ycap)
    if i%100 == 0:
        print(f"w1={w1}, w2={w2}, b={b}, Loss={L}")
    dw1 = -np.mean((y-ycap)*x1)
    dw2 = -np.mean((y-ycap)*x2)
    db = -np.mean(y-ycap)
    new_w1 = w1 - eta*dw1 
    new_w2 = w2 - eta*dw2
    new_b = b - eta*db 
    w1, w2, b = new_w1, new_w2, new_b
```
