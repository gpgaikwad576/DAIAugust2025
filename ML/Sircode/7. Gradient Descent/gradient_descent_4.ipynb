```python
import numpy as np
import pandas as pd
from sklearn.metrics import log_loss,  classification_report
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
```

```python
kyph = pd.read_csv("C:/Python/Cases/Kyphosis/Kyphosis.csv")
le = LabelEncoder()
kyph['Kyphosis'] = le.fit_transform( kyph['Kyphosis'] )
y = kyph['Kyphosis']
X = kyph.drop('Kyphosis', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, 
                                                    random_state=25)
mm = MinMaxScaler().set_output(transform='pandas')
X_trn_scl = mm.fit_transform(X_train)
X_tst_scl = mm.transform(X_test)
```

```python
sgd = SGDClassifier(random_state=25,loss='log_loss', learning_rate='adaptive', eta0=0.1)
sgd.fit(X_trn_scl, y_train)
y_pred_prob = sgd.predict_proba(X_tst_scl)
print(log_loss(y_test, y_pred_prob))
```

```python
y_pred = sgd.predict(X_tst_scl)
print( classification_report( y_test, y_pred ) )
```

##### HR Analytics

```python
hr = pd.read_csv("C:/Python/Cases/human-resources-analytics/HR_comma_sep.csv")
dum_hr = pd.get_dummies(hr, drop_first=True )
X = dum_hr.drop('left', axis=1)
y = dum_hr['left']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, 
                                                    random_state=25)
mm = MinMaxScaler().set_output(transform='pandas')
X_trn_scl = mm.fit_transform(X_train)
X_tst_scl = mm.transform(X_test)
```

```python
sgd = SGDClassifier(random_state=25,loss='log_loss')
sgd.fit(X_trn_scl, y_train)
y_pred_prob = sgd.predict_proba(X_tst_scl)
print(log_loss(y_test, y_pred_prob))
```

```python
y_pred = sgd.predict(X_tst_scl)
print( classification_report( y_test, y_pred ) )
```
