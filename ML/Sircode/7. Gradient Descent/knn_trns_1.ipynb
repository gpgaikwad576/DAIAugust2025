```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer, make_column_selector
```

```python
hr = pd.read_csv("C:/Python/Cases/human-resources-analytics/HR_comma_sep.csv")
X = hr.drop('left', axis=1)
y = hr['left']
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first').set_output(transform='pandas')
trns = make_column_transformer((ohe, make_column_selector(dtype_include=object) ), 
                               remainder='passthrough',
                               verbose_feature_names_out=False)
trns = trns.set_output(transform='pandas')
X_trns = trns.fit_transform(X)
scaler = StandardScaler().set_output(transform='pandas')
X_scaled = scaler.fit_transform(X_trns)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_scaled, y)
```

```python
tst = pd.read_csv("C:/Python/Cases/human-resources-analytics/tst_hr.csv")
tst_trns = trns.transform(tst)
tst_scaled = scaler.transform( tst_trns )
knn.predict( tst_scaled )
```

```python
from sklearn.pipeline import Pipeline
```

```python
trns = make_column_transformer((ohe, make_column_selector(dtype_include=object) ), 
                               remainder='passthrough',
                               verbose_feature_names_out=False)
scaler = StandardScaler()
knn = KNeighborsClassifier(n_neighbors=1)
pipe = Pipeline( [('OHE',trns),('SCL',scaler),('KNN',knn)] )
pipe.fit(X, y)
pipe.predict(tst)
```
