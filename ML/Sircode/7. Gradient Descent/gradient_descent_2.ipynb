```python
import numpy as np
import pandas as pd
from sklearn.metrics import log_loss, accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
kyph = pd.read_csv("C:/Python/Cases/Kyphosis/Kyphosis.csv")
le = LabelEncoder()
kyph['Kyphosis'] = le.fit_transform( kyph['Kyphosis'] )
y = kyph['Kyphosis']
X = kyph.drop('Kyphosis', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, 
                                                    random_state=25)
mm = MinMaxScaler().set_output(transform='pandas')
X_trn_scl = mm.fit_transform(X_train)
X_tst_scl = mm.transform(X_test)
X_trn_scl.shape
```

```python
x1, x2, x3 = X_trn_scl['Age'], X_trn_scl['Number'], X_trn_scl['Start']
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
eta = 0.1
b, w1, w2, w3 = 0.1, 0.1, 0.1, 0.1
for i in range(1000):
    ycap = sigmoid(w1*x1 + w2*x2 + w3*x3 + b)
    L = log_loss( y_train , ycap)
    if i%100 == 0:
        print(f"w1={w1}, w2={w2}, w3={w3} , b={b}, Loss={L}")
    dw1 = -np.mean((y_train-ycap)*x1)
    dw2 = -np.mean((y_train-ycap)*x2)
    dw3 = -np.mean((y_train-ycap)*x3)
    db = -np.mean(y_train-ycap)
    new_w1 = w1 - eta*dw1 
    new_w2 = w2 - eta*dw2
    new_w3 = w3 - eta*dw3
    new_b = b - eta*db 
    w1, w2, w3 ,b = new_w1, new_w2, new_w3, new_b
```

```python
w1, w2, w3, b
```

Equivalent code for `.predict_proba( )`

```python
y_pred_prob = sigmoid(w1*X_tst_scl['Age'] + w2*X_tst_scl['Number'] + w3*X_tst_scl['Start'] + b)
print("Log Loss on Test set:", log_loss(y_test, y_pred_prob))
```

```python
y_pred = np.where(y_pred_prob>0.5, 1, 0)
print(classification_report(y_test, y_pred))
```
