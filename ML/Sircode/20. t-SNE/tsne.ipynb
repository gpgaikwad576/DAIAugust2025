```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
wisconsin = pd.read_csv("C:/Python/Cases/Wisconsin/BreastCancer.csv", index_col=0)
y = wisconsin['Class']
X = wisconsin.drop('Class', axis=1)
```

```python
prcomp = PCA().set_output(transform='pandas')
X_pca = prcomp.fit_transform( X )
pc_data = X_pca.copy()
pc_data['Class'] = y
sns.scatterplot( data=pc_data, x='pca0',y='pca1', hue='Class' )
plt.title("Using PCA")
plt.show()
```

```python
tsne = TSNE(random_state=25).set_output(transform='pandas')
X_tsne = tsne.fit_transform(X)
tsne_data = X_tsne.copy()
tsne_data['Class'] = y
sns.scatterplot( data=tsne_data, x='tsne0',y='tsne1', hue='Class' )
plt.title("Using t-SNE")
plt.show()
```

Vehicle

```python
vehicle = pd.read_csv("C:/Python/Cases/Vehicle_Silhouettes/Vehicle.csv")
y = vehicle['Class']
X = vehicle.drop('Class', axis=1)
```

With scaling

```python
scaler = StandardScaler().set_output(transform='pandas')
X_scl = scaler.fit_transform(X)
tsne = TSNE(random_state=25,perplexity=45).set_output(transform='pandas')
X_tsne = tsne.fit_transform(X_scl)
tsne_data = X_tsne.copy()
tsne_data['Class'] = y
sns.scatterplot( data=tsne_data, x='tsne0',y='tsne1', hue='Class' )
plt.title("Using t-SNE")
plt.show()
```

```python
prcomp = PCA().set_output(transform='pandas')
X_scl = scaler.fit_transform(X)
X_pca = prcomp.fit_transform( X_scl )
pc_data = X_pca.copy()
pc_data['Class'] = y
sns.scatterplot( data=pc_data, x='pca0',y='pca1', hue='Class' )
plt.title("Using PCA")
plt.show()
```
