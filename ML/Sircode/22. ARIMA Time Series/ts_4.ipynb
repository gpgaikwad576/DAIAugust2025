```python
from statsmodels.tsa.stattools import adfuller
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import root_mean_squared_error
from ipywidgets import interact, widgets
from statsmodels.graphics.tsaplots import plot_acf
import os 
os.chdir("C:/Python/Datasets")
```

```python
df = pd.read_csv("FRED-NROUST.csv")
df['Value'].plot()
plt.show()
```

#### Augmented Dicky Fuller Test

```python
result = adfuller(df['Value'], maxlag=10)
print("P-Value =", result[1])
if result[1] < 0.05:
    print("Time Series is Stationary")
else:
    print("Time Series is not Stationary")
```

```python
ord_1_diff = df['Value'].diff() # does first order differencing
ord_1_diff = ord_1_diff.dropna()
result = adfuller(ord_1_diff, maxlag=10)
print("P-Value =", result[1])
if result[1] < 0.05:
    print("Time Series is Stationary")
else:
    print("Time Series is not Stationary")
```

```python
plot_acf(df['Value'], lags=30,alpha=None)
plt.show()
```
