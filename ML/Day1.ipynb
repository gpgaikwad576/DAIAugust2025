```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import root_mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r"D:\Datasets\Insure_auto.csv")
df
```

```python
x=df[['Home',"Automobile"]]
y=df['Operating_Cost']
x.shape,y.shape
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.intercept_,model.coef_[0]
```

```python
x=df['Home']
x=x.to_numpy().reshape(-1,1)
x.shape
```

## Transformation

```python
poly=PolynomialFeatures(degree=2,include_bias=False).set_output(transform='pandas')
x_poly=poly.fit_transform(x)
```

```python
x_poly
```

```python
model.fit(x_poly,y)
model.intercept_,model.coef_
```

```python
y_pred=model.intercept_+(model.coef_[0]*x)+(model.coef_[1]*x**2)
y_pred
```

```python
root_mean_squared_error(y,y_pred)
```

```python
r2_score(y,y_pred)
```

```python
y_model_pred=model.predict(x_poly)
y_model_pred
```

Boston Housing

```python
boston=pd.read_csv(r"D:\Datasets\Boston.csv")
x = boston.drop('medv',axis =1)
y = boston['medv']
```

```python
lr  = LinearRegression()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 25)
lr.fit(x,y)
y_pred = lr.predict(x_test)
mean_absolute_error(y_test,y_pred)
```

```python
poly = PolynomialFeatures(degree = 2).set_output(transform='pandas')
x_poly = poly.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x_poly,y,test_size = 0.3,random_state = 25)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
mean_absolute_error(y_test,y_pred)
```

```python
poly = PolynomialFeatures(degree = 3).set_output(transform='pandas')
x_poly = poly.fit_transform(x)
x_train,x_test,y_train,y_test = train_test_split(x_poly,y,test_size = 0.3,random_state = 25)
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
mean_absolute_error(y_test,y_pred)
```
