```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import lineStyles
```

```python
empname=['abc','def','hej']
salary=[100,200,1000]
plt.plot(empname,salary,linestyle='--',marker='o',markerfacecolor='g',linewidth=2,markeredgecolor='b',c='g',markeredgewidth=2)
```

```python
plt.bar(empname,salary,color=['red','blue','g'])
plt.xlabel('Salary')
plt.ylabel('name')
plt.title('Salary Graph',fontdict={'fontsize':16,'fontweight':'bold','color':'darkred'})
```

```python
fig,ax=plt.subplots()
name=['a'+str(i) for i in range(0,10)]
salary=[np.random.randint(1,100) for _ in range(0,10)]
plt.pie(salary,labels=name,explode=[0.05 for _ in range(0,10)],colors=[np.random.choice(['r','b','g','aquamarine','grey','magenta']) for _ in range(0,10)],autopct='%2.2f%%',radius=1.5)
plt.show()
```
