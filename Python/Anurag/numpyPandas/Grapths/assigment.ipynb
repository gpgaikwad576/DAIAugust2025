```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot  as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from collections import defaultdict
def extracttype(data):
    d=pd.DataFrame()
    data=data.split('\n')
    for line in data:
        type=line.split(' ')[2][1:-1]
        message=line.split('-')[-1]
        d=pd.concat([d,pd.DataFrame([{'type':type,'message':message}])])

    d.reset_index(inplace=True)
    d.drop('index',inplace=True,axis=1)

    return d
```

```python
with open(r"C:\Users\dai\Desktop\ecommerce_app_detailed.log",'r') as f:
    data=f.read()
    res=extracttype(data)

res
```
