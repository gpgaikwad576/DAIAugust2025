```python
top = Tk()
top.title('MY First Application')
top.geometry("400x200")

def find():
    num = int(e.get())
    print(f'square is {num*num}')
    # result = Label(top,text=f'square is {num*num}')
    # result.pack()
    result.config(text = f'Square is {num*num}')
    

l = Label(top,text='Enter the Number')
e = Entry(top)
b = Button(top,text='Find  Square ',command=find)
result = Label(top,text=f'Square is :')

l.pack()
e.pack()
b.pack()
result.pack()

top.mainloop()
```

```python
top = Tk()
top.title('MY First Application')
top.geometry("400x200")

def find_sum():
    num1 = int(e1.get())
    num2 = int(e2.get())
    print(f'Sum is {num1 + num2}')
    # result = Label(top,text=f'square is {num*num}')
    # result.pack()
    result1.config(text = f'Sum is {num1 + num2}')

def find_multiple():
    num1 = int(e1.get())
    num2 = int(e2.get())
    print(f'Sum is {num1 * num2}')
    # result = Label(top,text=f'square is {num*num}')
    # result.pack()
    result2.config(text = f'Multiple is {num1 * num2}')


def find_division():
    num1 = int(e1.get())
    num2 = int(e2.get())
    print(f'Sum is {num1 / num2}')
    # result = Label(top,text=f'square is {num*num}')
    # result.pack()
    result3.config(text = f'Division is {num1 / num2}')

    

l1 = Label(top,text='Enter the First Number')
e1 = Entry(top)

l2 = Label(top,text='Enter Second Number')
e2 = Entry(top)

b1 = Button(top,text='+',command=find_sum)
b2 = Button(top,text='* ',command=find_multiple)
b3 = Button(top,text='/',command=find_division)


result1 = Label(top,text=f'Sum is :')
result2 = Label(top,text=f'Multiplication is :')
result3 = Label(top,text=f'Division is :')

l1.place(x=50,y=50)
e1.place(x=50,y=70)

l2.place(x=50,y=90)
e2.place(x=50,y=110)

b1.place(x=50,y=130)
b2.place(x=70,y=130)
b3.place(x=90,y=130)

result1.pack()
result2.pack()
result3.pack()

top.mainloop()
```
