```python
from string import punctuation
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

sent = 'Hello friends! How are you? Welcome to Python Programming.'

# Tokenization
tokens = word_tokenize(sent)
print(tokens)

#Stopwords removal
words = [word for word in tokens if word not in set(stopwords.words('english'))]
print(words)

# puncuations removal
punctuations = string.punctuation
words_nop = [word for word in words if word if word not in punctuations ]
print(words_nop)
```
