```python
import pyttsx3  
from gtts import gTTS
from playsound import playsound             
```

```python
engine = pyttsx3.init()
```

```python
engine.setProperty('rate', 100)
engine.setProperty('volume', 0.9)

text = 'hello'
engine.say(text)                            
engine.runAndWait()
```

```python
text = 'Hello world welcome to the world of U & I'
tts = gTTS(text,lang='en')

tts.save('audio.mp3')
playsound('audio.mp3')
```
