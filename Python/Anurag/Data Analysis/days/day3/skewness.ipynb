```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\dai\Downloads\airline_dec_2008_50k (1).csv")
df
```

```python
df.info()
```

```python
data=df['Distance']

mean_dist=data.mean()
std_dist=data.std()
median_dist=data.median()

print('mean:',mean_dist)
print('median:',median_dist)
print('std:',std_dist)
```

```python
sns.histplot(data,kde=True,color='skyblue',bins=30)
plt.axvline(mean_dist,color='r',linestyle='--',label=f'mean:{mean_dist:.2f}')
plt.axvline(median_dist,color='orange',linestyle='--',label=f'median:{median_dist:.2f}')
plt.title('Skewness')
plt.xlabel('Distance')
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.show()
```
