## 4. Write a Python program to draw line charts of the financial data of Alphabet


```python
df=pd.read_csv(r"C:\Users\dai\Desktop\fdata.csv.txt")
df["Date"]=pd.to_datetime(df["Date"],dayfirst=False)
df["Date"]=df["Date"].dt.strftime("%d-\n%B-\n%Y")
plt.figure(figsize=(10,8))
plt.plot(df["Date"],df["Open"])
plt.plot(df["Date"],df["High"])
plt.plot(df["Date"],df["Low"])
plt.plot(df["Date"],df["Close"])
plt.legend(df.select_dtypes(include=['number']).columns)
plt.show()
```

## 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python
x=[10,15,20,25,30]
y1=[20,30,40,20,0]
y2=[40,20,0,15,30]

plt.plot(x,y1,color='b')
plt.xticks(ticks=x,labels=[10,15,20,25,30])
plt.plot(x,y2,color='green')
```

## 6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
x=[10,15,20,25,30]
y1=[20,30,40,20,0]
y2=[40,20,0,15,30]

plt.plot(x,y1,color='b',linewidth=3,label='Line 1 - width-3')
plt.xticks(ticks=x,labels=[10,15,20,25,30])
plt.plot(x,y2,color='r',linewidth=5,label='Line2 - width 5')
plt.legend()
plt.show()
```

## 7. Write a Python program to plot two or more lines with different styles.

```python
x=[10,15,20,25,30]
y1=[20,30,40,20,0]
y2=[40,20,0,15,30]

plt.plot(x,y1,color='b',linewidth=1,linestyle='dotted',label="line1-dotted")
plt.xticks(ticks=x,labels=[10,15,20,25,30])
plt.plot(x,y2,color='r',linewidth=3,linestyle="--",label="line2-dashed")
plt.show()
```

## 9. Write a Python program to display the current axis limits values and set new

```python
x=[0,20,40,60]
y=[0,50,100,150]
plt.plot(x,y)
plt.xlim(0,100)
plt.ylim(0,200)
plt.xticks(ticks=np.linspace(0,100,6))
plt.yticks(np.linspace(0,200,5))
plt.title('Draw a line')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## 10. Write a Python program to plot quantities which have an x and y position.

```python
data1 = [(4, 7), (10, 20), (15, 25), (20, 45), (30, 44), (33, 45), (20, 12)]
data2 = [(4, 3), (10, 15), (5, 12), (7, 4), (3, 44), (33, 50), (12, 20)]
x1, y1 = zip(*data1)
x2, y2 = zip(*data2)
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1,marker='o',edgecolors='red',c='red',label='Data 1 (Circles)')
plt.scatter(x2, y2,marker='*',edgecolors='blue',c='blue',s=150,label='Data 2 (Stars)')
plt.legend('upper right')
plt.show()
```

## 15. Write a Python program to create multiple plots.


```python
plt.figure(figsize=(10,6))
ax=plt.subplot2grid((2,4),(0,0),colspan=3)
for i in range(4):
    ax=plt.subplot2grid((2,4),(1,i))
plt.axis('off')
plt.show()
```
