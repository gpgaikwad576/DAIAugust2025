2. Create a data frame with at least 5 columns and 10 rows. There must be some missing
values in it. If the integer column contains missing values replace it with some statistical
approach. Also, if there is any missing values in String column, drop the rows using some
looping function.

```python
data = {
    'ProductID': [101, 102, 103, 104, 105, 106, 105, 108, 109, 110],
    'Price_USD': [19.99, 25.50, 15.00, np.nan, 30.25, 12.50, 45.00, 22.00, 17.75, 50.00],
    'Stock_Count': [50, 120, np.nan, 75, np.nan, 90, 15, 180, np.nan, 60], # Integer column with missing values
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Home', 'Clothing', np.nan, 'Electronics', 'Clothing', np.nan, 'Electronics'],
    'Rating': [4.5, 3.9, np.nan, 4.8, 4.2, 3.5, 4.9, 4.1, 3.0, 4.7]
}

df=pd.DataFrame(data)
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
index_value=[]
for index,data in df.iterrows():
    if sum([1 if pd.isnull(x) else 0 for x in data])>0:
        index_value.append(index)
df.drop(index_value)
```

3. Create a data frame with at least 5 columns and 10 rows. Each column must contain only
integer elements only. Column names must be Abilash, Ankit, Ashok, Asif, Anjaan. Now
apply a function where we want to club together the data based on any row or column.
Group the series using a mapper or by a series of columns.m

```python
def data_sum(rows):
    return sum([x for x in rows])
```

```python
np.random.seed(42)
data = {
    'Abilash': np.random.randint(10, 100, size=10),
    'Ankit': np.random.randint(50, 150, size=10),
    'Ashok': np.random.randint(1, 50, size=10),
    'Asif': np.random.randint(200, 300, size=10),
    'Anjaan': np.random.randint(5, 25, size=10)
}
df=pd.DataFrame(data)
df['Row_Total']=df.apply(data_sum,axis=1)
# pd.concat([df,pd.DataFrame(df.apply(data_sum,axis=0),index=False)],axis=0)
pd.concat([df,df.sum(axis=0).to_frame().T.rename(index={0: 'Col_Total'})])
```

5. Age Demographics
o The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
 Purchase Count
 Average Purchase Price
 Total Purchase Value
 Average Purchase Total per Person by Age Group

```python
df['Age_Group']=pd.cut(df['Age'],labels=["<10","10-20",'20-30','30+'],bins=[0,10,20,30,np.inf],right=True,include_lowest=True)
df.head()
```

```python
df.groupby('Age_Group',observed=True)['Purchase ID'].count()
```

```python
df.groupby('Age_Group',observed=True)['Price'].mean()
```

```python
df.groupby('Age_Group',observed=True)['Price'].sum()
```

```python
x_data=df.groupby('Age_Group',observed=True)['Price'].sum().to_frame(name="Total Purchase Price")
x_data['Average_Price_Per_Gender']=x_data['Total Purchase Price']/df.groupby("Age_Group",observed=True).nunique()['SN']
x_data
```
