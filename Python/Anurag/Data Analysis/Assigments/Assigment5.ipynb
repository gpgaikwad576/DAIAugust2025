## 14  Write a Python program to create a stacked bar plot with error bars.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

p1 = plt.bar(ind, menMeans, width, yerr=menStd, color='red')
p2 = plt.bar(ind, womenMeans, width,
bottom=menMeans, yerr=womenStd, color='green')

plt.ylabel('Scores')
plt.xlabel('Groups')
plt.title('Scores by group\n' + 'and gender')
plt.xticks(ind, ('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

## 15. Write a Python program to create a horizontal bar chart with differently ordered colors.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Sample Data (Scores for each student by subject)
# Note: The sum of scores for Student-1 and Student-3 is around 250,
# and Student-2 is also around 250 in the plot.
data = {
    'Student': ['Student-1', 'Student-2', 'Student-3'],
    'Language': [75, 71, 75],
    'Math': [96, 92, 89],
    'Science': [88, 95, 90]
}

df = pd.DataFrame(data)

# 2. Set up the plotting parameters
# Reverse the student order for plotting since barh plots from bottom up
students = df['Student'][::-1]
language_scores = df['Language'][::-1]
math_scores = df['Math'][::-1]
science_scores = df['Science'][::-1]

# Define colors for the stacked segments
colors = {
    'Language': 'red',
    'Math': 'green',
    'Science': 'blue'
}

# 3. Create the stacked bar chart

plt.figure(figsize=(10, 5))

# --- Plot the first segment (Language) ---
# 'left' is set to 0, meaning it starts at the beginning of the bar.
plt.barh(
    students,
    language_scores,
    color=colors['Language'],
    edgecolor='white',
    label='Language'
)

# --- Plot the second segment (Math) ---
# 'left' is set to the sum of the previous segment (Language),
# so the Math segment starts where Language ends.
language_cumulative = language_scores.cumsum() - language_scores
plt.barh(
    students,
    math_scores,
    left=language_cumulative,
    color=colors['Math'],
    edgecolor='white',
    label='Math'
)

# --- Plot the third segment (Science) ---
# 'left' is set to the sum of the two previous segments (Language + Math).
math_cumulative = language_cumulative + math_scores
plt.barh(
    students,
    science_scores,
    left=math_cumulative,
    color=colors['Science'],
    edgecolor='white',
    label='Science'
)

# 4. Final Touches
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()
plt.show()
```

## 16. Write a Python program to create stack bar plot and add label to each section.

```python
import numpy as np
import matplotlib.pyplot as plt

people = ('G1','G2','G3','G4','G5','G6','G7','G8')
segments = 4

data = [[  3.40022085,   7.70632498,   6.4097905,   10.51648577,   7.5330039,
    7.1123587,   12.77792868,   3.44773477],
 [ 11.24811149,   5.03778215,   6.65808464,  12.32220677,   7.45964195,
    6.79685302,   7.24578743,   3.69371847],
 [  3.94253354,   4.74763549,  11.73529246,   4.6465543,   12.9952182,
    4.63832778,  11.16849999,   8.56883433],
 [  4.24409799,  12.71746612,  11.3772169,    9.00514257,  10.47084185,
   10.97567589,   3.98287652,   8.80552122]]
percentages = (np.random.randint(5,20, (len(people), segments)))
y_pos = np.arange(len(people))

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

colors ='rgwm'
patch_handles = []
left = np.zeros(len(people))
for i, d in enumerate(data):
    patch_handles.append(ax.barh(y_pos, d,
      color=colors[i%len(colors)], align='center',
      left=left))
    left += d

for j in range(len(patch_handles)):
    for i, patch in enumerate(patch_handles[j].get_children()):
        bl = patch.get_xy()
        x = 0.5*patch.get_width() + bl[0]
        y = 0.5*patch.get_height() + bl[1]
        ax.text(x,y, "%d%%" % (percentages[i,j]), ha='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Scores')
plt.show()
```

## 17. Write a Python program to add textures (black and white) to bars and wedges.

```python
fig = plt.figure()
patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*","x", "o", "O" ]

ax = fig.add_subplot(111)
for i in range(len(patterns)):
    ax.bar(i, 3, color='white', edgecolor='black', hatch=patterns[i])

plt.show()
```
