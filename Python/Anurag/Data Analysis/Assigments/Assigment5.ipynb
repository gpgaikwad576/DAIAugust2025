```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Write a Python programming to display a bar chart of the popularity of programming Languages.

```python
Programming_languages='Java  Python  PHP  JavaScript  C#  C++'.split()
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(Programming_languages,Popularity,color='b')
plt.xlabel('languages',ha='left')
plt.xlim(0)
plt.yticks(np.linspace(0,25,6))
plt.ylabel('Popularity')
plt.title('popularity of Programming Language')
plt.show()
```

## 2. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.

```python
Programming_languages='Java  Python  PHP  JavaScript  C#  C++'.split()
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax=sns.barplot(y=Programming_languages,x=Popularity,color='green',order=Programming_languages[::-1])
plt.xlabel('Popularity')
plt.ylabel('Languages')
plt.title('popularity of Programming Language')
plt.show()
```

## 4. Write a Python programming to display a bar chart of the popularity of programming Languages. Use different color for each bar.

```python
Programming_languages='Java  Python  PHP  JavaScript  C#  C++'.split()
Popularity= [22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax=sns.barplot(x=Programming_languages,y=Popularity,order=Programming_languages,palette=['r','white','g','b','yellow','skyblue'],hue=Programming_languages)
plt.xlabel('Popularity')
plt.ylabel('Languages')
plt.title('popularity of Programming Language')
plt.show()
```

## 5. Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity (float value).

```python
Programming_languages='Java  Python  PHP  JavaScript  C#  C++'.split()
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
ax=sns.barplot(x=Programming_languages,y=Popularity,order=Programming_languages,palette=['r','white','g','b','yellow','skyblue'],hue=Programming_languages)
for container in ax.containers:
    ax.bar_label(container)
plt.xlabel('Popularity')
plt.ylabel('Languages')
plt.title('popularity of Programming Language')
plt.show()
```

## 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.

```python
Means_men=[22, 30, 35, 35, 26]
Means_women= [25, 32, 30, 35, 29]
data=[]
men=[]
women=[]
gender=[]
for i in range(0,len(Means_women)):
   data.append(Means_men[i])
   men.append(Means_men[i])
   data.append(Means_women[i])
   women.append(Means_women[i])
   gender.append("Men")
   gender.append('Women')

df=pd.DataFrame({"Sex":gender,"men":men,'women':women})
print(data,men,women)
df

```
