## Edge Detection

## Sobel

x = [-1,0,1],
    [-2,0,2],
    [-1,0,1]

y=  [-1,-2,-1],
    [0,0,0],
    [1,2,1]  

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
```

```python
sobelx.dtype
```

```python
final_sobel = cv2.add(sobelx,sobely)
```

```python
thresh,binary_sobel = cv2.threshold(final_sobel,60,55,cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(16,9))
plt.subplot(141)
plt.title('Original')
plt.imshow(img,cmap='gray')

plt.subplot(142)
plt.title('Only sobel x')
thresh,binary_sobel_x = cv2.threshold(sobelx,60,55,cv2.THRESH_BINARY)
plt.imshow(binary_sobel_x,cmap='gray')

plt.subplot(143)
plt.title('Only sobel y')
thresh,binary_sobel_y = cv2.threshold(sobely,60,55,cv2.THRESH_BINARY)
plt.imshow(binary_sobel_y,cmap='gray')

plt.subplot(144)
plt.title('Edges Sobel x+y')
plt.imshow(binary_sobel,cmap='gray')
```

## Canny Edge Detection

```python
img = cv2.imread(r'datasets\image_dataset\images\lena.png',0)

edges = cv2.Canny(img,100,200)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title(f'Canny Edge')
plt.axis('off')

plt.show()
```
