```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

def nothing():
    pass

trackbar_window = 'Trackbars'


image  = cv.imread('datasets\image_dataset\images\cat.jpg')
image = cv.resize(image,(400,400))


hsv = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cv.namedWindow(trackbar_window)

cv.createTrackbar('H Min', trackbar_window, 0, 255, nothing)
cv.createTrackbar('S Min', trackbar_window, 0, 255, nothing)
cv.createTrackbar('V Min', trackbar_window, 0, 255, nothing)

# Create trackbars for HSV maximum values
cv.createTrackbar('H Max', trackbar_window, 179, 179, nothing)
cv.createTrackbar('S Max', trackbar_window, 255, 255, nothing)
cv.createTrackbar('V Max', trackbar_window, 255, 255, nothing)

while True:
    h_min = cv.getTrackbarPos('H Min',trackbar_window)
    s_min = cv.getTrackbarPos('S Min',trackbar_window)
    v_min = cv.getTrackbarPos('V Min',trackbar_window)

    h_max = cv.getTrackbarPos('H Max',trackbar_window)
    s_max = cv.getTrackbarPos('S Max',trackbar_window)
    v_max = cv.getTrackbarPos('V Max',trackbar_window)


    lower_bound = np.array([h_min, s_min, v_min])
    upper_bound = np.array([h_max, s_max, v_max])

    mask = cv.inRange(hsv, lower_bound, upper_bound)

    result = cv.bitwise_and(image, image, mask=mask)

    cv.imshow('Original Image', image)
    cv.imshow('Masked Result (Color Filter)', result)
    cv.imshow('Mask Only', mask)

    key = cv.waitKey(1) & 0xFF
    if key == ord('q'):
        break
```
