```python
import re 
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
```

## **Question1**

```python
sent = 'They told that their ages are 23 , 25 and 29 respectively.'
```

### **1. Find the average of ages.**

```python
numbers = re.findall(r'\d+',sent)
numbers = list(map(int,numbers))
sum(numbers)/len(numbers)
```

```python
numbers = [word for word in sent.split(' ') if word.isdigit()]
numbers = list(map(int,numbers))

sum(numbers)/len(numbers)
```

## 2.Count total No of Words Starts with 't'/'T'

```python
words = [word for word in sent.split() if word.lower().startswith('t')]

print('Total number of words starts with T/t is:',len(words))
```

## 3.Count total punctuaion symbols.

```python
import string

len([word for word in sent if word in string.punctuation])
```

### **4. Print the words in capital letter which ends with 'd'**

```python
[word.upper() for word in sent.split() if word.endswith('d')]
```

### **5.Merge all the words and print (no punctuations)**

```python
"".join([word for word in sent if word not in string.punctuation])
```
