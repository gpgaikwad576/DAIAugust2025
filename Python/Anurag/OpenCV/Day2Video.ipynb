```python
cap = cv2.VideoCapture(r'datasets\image_dataset\images\robots.mp4')
if not cap.isOpened():
    print("Error: Could not open video file.")
    exit()


output_dir = 'datasets/image_dataset/images/output_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
    print('Created directory Successfully: ',output_dir)

cnt=1

while cap.isOpened():
    _,frame = cap.read()

    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    

    if _ :
        cv2.imshow('Video Frame',frame)
        cv2.imshow('gray video',gray)
        
        cv2.imwrite(f'datasets/image_dataset/images/output_images/frame{cnt}.png',frame)
        cnt+=1

        if cv2.waitKey(10)==27:
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('D:\DAIAugust2025\Python\Anurag\OpenCV\datasets\image_dataset\images\dog.jpg')
cv2.imshow('my Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
quater_width,quater_height = img.shape[1]//4,img.shape[0]//4
```

```python
# define translation matrix
T = np.float32([[1,0,quater_width],[0,1,quater_height]]) 
```

```python
T
```

```python
translated_img = cv2.warpAffine(img,T,(img.shape[1],img.shape[0]))
```

```python
cv2.imshow('Translated Image',translated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
translated_img = cv2.warpAffine(img,T,(600,600))
cv2.imshow('Translated Image',translated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
T = np.float32([[1,0,150],[0,1,75]])
translated_img = cv2.warpAffine(img,T,(img.shape[1],img.shape[0]))
cv2.imshow('Translated_image',translated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
T = np.float32([[1,0,-100],[0,1,-75]])
translated_img = cv2.warpAffine(img,T,(600,600))
cv2.imshow('Translated_image',translated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_90 = cv2.rotate(img,cv2.ROTATE_90_CLOCKWISE)
img_90a = cv2.rotate(img,cv2.ROTATE_90_COUNTERCLOCKWISE)
img_180 = cv2.rotate(img,cv2.ROTATE_180)
```

```python
cv2.imshow('90',img_90)
cv2.imshow('90_anti',img_90a)
cv2.imshow('180',img_180)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
center = img.shape[1]//2,img.shape[0]//2
```
