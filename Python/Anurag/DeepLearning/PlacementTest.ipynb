```python
def unique_len(str1):
    us=''
    for i in str1:
        if i not in us:
            us+=i
    return f'{len(us)} "{us}"'
print(unique_len("abcabcbb"))
print(unique_len('bbbbb'))
```

```python
def unique_len(str1):
    us=''
    l=0
    for i in str1:
        if i not in us:
            us+=i
            l+=1
    return f'{l} "{us}"'
print(unique_len("abcabcbb"))
print(unique_len('bbbbb'))
```
